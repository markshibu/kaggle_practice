```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
reviews = pd.read_csv('winemag-data_first150k.csv',index_col=0)
reviews.head()
```

```python
reviews['points'].value_counts().sort_index().plot.bar()
```

```python
reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(8, 4),
    color='mediumvioletred',
    fontsize = 20,
    title='Rankings Given by Wine Magazine',
)
```

### pandas doesn't give us an easy way of adjusting the title size, but anything that you build in pandas can be built using matplotlib directly.

```python
ax = reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(8, 4),
    color='mediumvioletred',
    fontsize = 16,
)
ax.set_title('Rankings Given by Wine Magazine',fontsize=20)
```

### sns.despine - turn off the ugly black border.

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(8,4),
    color='mediumvioletred',
    fontsize=16
)
ax.set_title("Rankings Given by Wine Magazine", fontsize=20)
sns.despine(bottom=True, left=False)
```
