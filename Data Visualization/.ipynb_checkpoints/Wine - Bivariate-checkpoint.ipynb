```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
reviews = pd.read_csv('winemag-data_first150k.csv',index_col=0)
reviews.head()
```

```python
# need to downsample
reviews[reviews['price'] < 100].plot.scatter(x='price', y='points',figsize=(12,8),s=1)
```

```python
reviews[reviews['price'] < 100].sample(300).plot.scatter(x='price', y='points',figsize=(12,8),s=1)
```

```python
reviews[reviews['price']<100].plot.hexbin(x='price',y='points',gridsize=20,yticks=np.arange(80,101,2),figsize=(8,6.5),sharex=False)
```
