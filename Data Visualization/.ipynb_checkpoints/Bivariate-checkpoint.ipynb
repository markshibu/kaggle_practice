```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Pokemon

```python
pd.set_option('max_columns', None)
pokemon = pd.read_csv("Pokemon-bivariate.csv",index_col=0)
pokemon.head(3)
```

```python
pokemon.plot.scatter(x='Attack',y='Defense')
```

```python
pokemon.plot.hexbin(x='Attack',y='Defense',gridsize=20,sharex=False)
```

```python
pokemon_stats_legendary = pokemon.groupby(['Legendary', 'Generation']).mean()[['Attack', 'Defense']]
pokemon_stats_legendary
```

```python
pokemon_stats_legendary.plot.bar()
```

```python
pokemon_stats_legendary.plot.bar(stacked=True)
```

```python
pokemon_stats_by_generation = pokemon.groupby('Generation').mean()[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
pokemon_stats_by_generation
```

```python
pokemon_stats_by_generation.plot.line()
```

# Wine

```python
reviews = pd.read_csv('winemag-data_first150k.csv',index_col=0)
reviews.head()
```

```python
# need to downsample
reviews[reviews['price'] < 100].plot.scatter(x='price', 
                                             y='points',
                                             figsize=(12,8),
                                             s=1)
```

```python
reviews[reviews['price'] < 100].sample(300).plot.scatter(x='price', 
                                                         y='points',
                                                         figsize=(12,8),
                                                         s=1)
```

```python
reviews[reviews['price']<100].plot.hexbin(x='price',
                                          y='points',
                                          gridsize=20,
                                          yticks=np.arange(80,101,2),
                                          figsize=(8,6.5),
                                          sharex=False)
```
