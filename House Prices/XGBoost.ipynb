```python
import pandas as pd
from sklearn.preprocessing import Imputer
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('train.csv')
```

```python
# Drop all rows with missing data on column 'SalePrice'
data = data.dropna(axis=0, how='any',subset=['SalePrice'])
```

```python
y = data.SalePrice
X = data.drop(['SalePrice'],axis=1).select_dtypes(exclude=['object'])
```

## Handling Missing Data

```python
def missing_data_marker(df):
    cols_with_missing = (col for col in df.columns if df[col].isnull().any())
    for col in cols_with_missing:
        df[col+'_was_missing'] = df[col].isnull()
    return df
```

```python
from sklearn.preprocessing import Imputer

X = missing_data_marker(X)
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.25)
```

```python
my_imputer = Imputer()
train_X = my_imputer.fit_transform(train_X)
test_X = my_imputer.transform(test_X)
```

## Fit & Evaluate

```python
from xgboost import XGBRegressor

my_model = XGBRegressor()
my_model.fit(train_X, train_y, verbose=False)
```

```python
from sklearn.metrics import mean_absolute_error

predictions = my_model.predict(test_X)
print("Mean Absolute Error : " + str(mean_absolute_error(predictions, test_y)))
```

## Model Tuning

#### n_estimators & early_stopping_rounds

#### Learning rate

```python
my_model = XGBRegressor(n_estimators=3000, learning_rate=0.01, n_jobs=4)
early_stopping_rounds = 5
my_model.fit(train_X, train_y, early_stopping_rounds=early_stopping_rounds, 
             eval_set=[(test_X, test_y)], verbose=True)
```

```python
rst = my_model.evals_result()
n_estimators = len(rst['validation_0']['rmse'])-early_stop_rounds
n_estimators
```

```python
predictions = my_model.predict(test_X)
print("Mean Absolute Error : " + str(mean_absolute_error(predictions, test_y)))
```

#### n_jobs
set the parameter n_jobs equal to the number of cores

### Retrain with all data

```python
my_model = XGBRegressor(n_estimators=n_estimators, learning_rate=0.01,n_jobs=4)
my_model.fit(train_X,train_y)
```

```python
predictions = my_model.predict(test_X)
print("Mean Absolute Error : " + str(mean_absolute_error(predictions, test_y)))
```

### Pipelines & Cross Validation

```python
from sklearn.pipeline import make_pipeline
my_pipeline = make_pipeline(Imputer(), XGBRegressor(n_estimators=n_estimators, learning_rate=0.01,n_jobs=4))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(my_pipeline, X, y, scoring='neg_mean_absolute_error',cv=5)
print('CrosValidation MAE : %2f' %(-1 * scores.mean()))
```

```python
my_pipeline.fit(train_X, train_y)
predictions = my_pipeline.predict(test_X)
print("MAE : " + str(mean_absolute_error(predictions, test_y)))
```
